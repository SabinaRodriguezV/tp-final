```python
print("A pesar de no trabajar con outliers, realizamos un cuadro de dispersion, sabiendo que la dispersion nos dara cero")
```

```python
df['Unificada'] = pd.to_numeric(df['Unificada'], errors='coerce')
```

```python
df.dtypes
```

```python
Q1 = df["Unificada"].quantile(0.25)
Q3 = df["Unificada"].quantile(0.75)
iqr = Q3 - Q1

umbral_superior = Q3 + 1.5 * iqr
umbral_inferior = Q1 - 1.5 * iqr

outliers = df[(df["Unificada"] > umbral_superior) | (df["Unificada"] < umbral_inferior)]
print("cantidad de outliers en Unificada", len(outliers))
plt.scatter(df.index,df["Unificada"],label="datos")
plt.scatter(outliers.index,outliers["Unificada"],label="o")
plt.title("Dispersion en outliers Unificada")
plt.show ()
```
